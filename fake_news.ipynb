```
import pandas as pd
dev = pd.read_csv('input/dev.tsv',sep='\t',header = None)
train = pd.read_csv('input/train.tsv',sep='\t',header = None)
```

```
train=train.drop([4],axis = 1)
dev=dev.drop([4],axis = 1)
```

```
train_df_bert = pd.DataFrame({
    'id':range(len(train)),
    'label':train[1],
    'alpha':['a']*train.shape[0],
    'text': train[3].replace(r'\n', ' ', regex=True)
})
```

```
def bin_label(df,label):
  lis = list(set([0,1,2,3,4,5]) - set([label]))
  df['label'] = df['label'].replace(lis, 'zero')
  df['label'] = df['label'].replace([label], 1)
  df['label'] = df['label'].replace(['zero'], 0)
  return df
  
```
